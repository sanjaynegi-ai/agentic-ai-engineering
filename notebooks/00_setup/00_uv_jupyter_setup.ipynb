# uv and Jupyter setup

## Learning goals

- Create a reproducible Python 3.12 environment
- Understand how `uv`, `.venv`, Python, and Jupyter fit together
- Register, select, inspect, and remove a Jupyter kernel


`uv` manages this project's Python version and dependencies. It creates an isolated environment in `.venv`, records dependencies in `pyproject.toml`, and locks exact resolved versions in `uv.lock`.

Use `uv run <command>` when you want `uv` to run a command with the project environment automatically. Activating `.venv` is optional, but it is convenient when you plan to run several Python commands in the same terminal session.


## Mental model

```mermaid
flowchart LR
  A[pyproject.toml + uv.lock] -->|uv sync / uv add| B[.venv]
  B --> C[Project Python 3.12]
  C --> D[Installed packages]
  C -->|ipykernel install| E[Jupyter kernel specification]
  E -->|selected in notebook| F[Jupyter notebook]
  F -->|runs cells with| C
```

The diagram shows two connected layers:

1. **Project environment:** `uv` reads `pyproject.toml` and `uv.lock`, then creates or updates `.venv`. That folder contains the project's Python interpreter and installed packages.
2. **Jupyter connection:** registering an `ipykernel` creates a small kernel specification that points Jupyter to the Python interpreter in `.venv`. Selecting **Agentic AI Engineering** tells the notebook to run every code cell with that interpreter and its packages.

> Registering a kernel does not create another virtual environment or copy your packages. It gives Jupyter a named pointer to the existing project environment.

## Setup commands

Choose the section for your terminal. For this existing course repository, run `uv sync` rather than `uv init`. Use `uv init` only when creating a new project in an empty folder.

### Windows PowerShell

```powershell
# New project only
uv init --python 3.12
uv add python-dotenv ipykernel

# Existing course repository
uv sync

# Activate the environment (optional)
.\.venv\Scripts\Activate.ps1

# Register and inspect the Jupyter kernel
uv run python -m ipykernel install --user --name agentic-ai-engineering --display-name "Agentic AI Engineering"
uv run jupyter kernelspec list
```

### Windows Command Prompt (cmd.exe)

```bat
REM New project only
uv init --python 3.12
uv add python-dotenv ipykernel

REM Existing course repository
uv sync

REM Activate the environment (optional)
.venv\Scripts\activate

REM Register and inspect the Jupyter kernel
uv run python -m ipykernel install --user --name agentic-ai-engineering --display-name "Agentic AI Engineering"
uv run jupyter kernelspec list
```

> In Command Prompt, `.venv\Scripts\activate` resolves to the `activate.bat` script. It does not work in PowerShell, which uses `Activate.ps1`.

### macOS or Linux (zsh/bash)

```bash
# New project only
uv init --python 3.12
uv add python-dotenv ipykernel

# Existing course repository
uv sync

# Activate the environment (optional)
source .venv/bin/activate

# Register and inspect the Jupyter kernel
uv run python -m ipykernel install --user --name agentic-ai-engineering --display-name "Agentic AI Engineering"
uv run jupyter kernelspec list
```

Activation makes `python` and installed command-line tools resolve from `.venv` in the current terminal. It is optional when you run commands through `uv run`.

The internal kernel name is `agentic-ai-engineering`; the display name is what you see in Jupyter or VS Code. After registration, select **Agentic AI Engineering** as this notebook's kernel.

### Unregister the kernel later (all platforms; do not run now)

```text
uv run jupyter kernelspec uninstall agentic-ai-engineering
```

> Unregistering removes only the Jupyter kernel specification. It does not delete `.venv`, uninstall packages, or change `pyproject.toml`.


## Verify the selected kernel

Run the Python cell above and inspect the first printed line. It should end with `.venv\Scripts\python.exe` on Windows or `.venv/bin/python` on macOS/Linux. If it points to another Python installation, select **Agentic AI Engineering** from the notebook's kernel picker and restart the kernel.
